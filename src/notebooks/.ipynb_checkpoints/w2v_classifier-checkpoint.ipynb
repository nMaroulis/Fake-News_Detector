```python
%matplotlib inline
import pandas as pd
import numpy as np
import sys, io, os, errno, fileinput, csv
import collections as cl
from os import path
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
import matplotlib as plt
import seaborn as sb
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import cross_val_score
import tensorflow as tf
from tensorflow import keras as kr
import itertools
import warnings
from sklearn.model_selection import StratifiedKFold
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import cross_val_score

trainset_file = "../../Datasets/train_1.csv"
train_df = pd.read_csv(trainset_file,  sep=',')


count_vect = CountVectorizer(stop_words='english', lowercase=True, ngram_range=(1, 2))
tfidf_transformer = TfidfTransformer(smooth_idf=False)
# svd = TruncatedSVD(n_components=16384)
train_df = train_df.dropna(subset=['text'])
```

__Load Google W2V Model__

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import ExtraTreesClassifier
from gensim.models import Word2Vec, KeyedVectors
import nltk

class TfidfEmbeddingVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        self.word2weight = None
        self.dim = len(next(iter(word2vec.items())))

    def fit(self, X, y):
        tfidf = TfidfVectorizer(analyzer=lambda x: x)
        tfidf.fit(X)
        # if a word was never seen - it must be at least as infrequent
        # as any of the known words - so the default idf is the max of 
        # known idf's
        max_idf = max(tfidf.idf_)
        self.word2weight = cl.defaultdict(
            lambda: max_idf,
            [(w, tfidf.idf_[i]) for w, i in tfidf.vocabulary_.items()])

        return self

    def transform(self, X):
        return np.array([
                np.mean([self.word2vec[w] * self.word2weight[w]
                         for w in words if w in self.word2vec] or
                        [np.zeros(self.dim)], axis=0)
                for words in X
            ])
    
num_features = 400      # Word vector dimensionality           
min_word_count = 140  # Minimum word count
num_workers = 4        # Number of threads to run in parallel
context = 10            # Context window size                                                            
downsampling = 1e-3    # Downsample setting for frequent words

model = KeyedVectors.load_word2vec_format("../../models/GoogleNews-vectors-negative300.bin", binary=True)
```

__Create TfIdf Word Embedding__

```python
model.vectors.shape
w2v = dict(zip(model.index2word, model.vectors)) # wv.index2word wv.syn0
tfidf_w2v = TfidfEmbeddingVectorizer(w2v)
```

__Normalization Functions__

```python
def w2v_Transformer(x,y):
    new_x = np.empty((0,300), np.float32) # standard Google vector size
    new_y = np.empty((0, 1), np.float64)
    
    for i in range(0,len(x)):

        if x[i].size == 300:
            new_x =  np.append(new_x, x[i].reshape(1,300), axis=0)
            new_y =  np.append(new_y, y[i].reshape(1,1), axis=0)
    return new_x, new_y.ravel()
```

__Corpus to W2V__

```python
_ = tfidf_w2v.fit(train_df['text'],train_df['label'].values)
w2v_matrix = _.transform(train_df['text'])


X, Y = w2v_Transformer(w2v_matrix, train_df['label'].values)
```

__Classification with  SVM__

```python
seed = 7
np.random.seed(seed)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
cvscores = []

svm_model = svm.SVC(kernel='linear', C=0.92, decision_function_shape='ovo')

i = 0
for train, test in kfold.split(X, Y):
    
    _ = svm_model.fit(X[train],Y[train]) # train

    i += 1
    predicted = svm_model.predict(X[test])
    print('acc ', str( np.mean(predicted == Y[test])))
```

__Tensorflow DNN Classifier Evaluation__

```python
seed = 7
np.random.seed(seed)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
cvscores = []

layer1 = [ 8 , 16 , 32 , 64 , 128 , 128 , 64 , 32 , 16 , 8 ]
layer2 = [ 16 , 16 , 32 , 16 , 16 , 8 , 16 , 32 , 64 , 128 ]
l_relu= 'relu'#kr.layers.LeakyReLU(alpha=0.3)
# 64 64 32
kmodel = kr.models.Sequential()
kmodel.add(kr.layers.Dense(32, input_shape=X[0].shape, activation=l_relu))
kmodel.add(kr.layers.Dense(32, activation=l_relu))
kmodel.add(kr.layers.Dense(16, activation=l_relu))
kmodel.add(kr.layers.Dense(1, activation='sigmoid'))
kmodel.summary()
#adam_opt = kr.optimizers.Adam(lr=0.0003, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
kmodel.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')

# kmodel.fit(X,Y, epochs=1000, batch_size=10, verbose=1, shuffle=True, class_weight=None, sample_weight=None) # train

i = 0
for train, test in kfold.split(X, Y):
    
    kmodel.fit(X[train],Y[train], epochs=100, batch_size=128, verbose=0, shuffle=True, class_weight=None, sample_weight=None) # train

    i += 1
    scores = kmodel.evaluate(X[test], Y[test], verbose=0)
    
    print("<%d> %s: %.2f%%" %  (i,kmodel.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)

    
print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
```

__Tensorflow DNN Classifier Final Model Build__

```python
fmodel = kr.models.Sequential()
fmodel.add(kr.layers.Dense(32, input_shape=X[0].shape, activation='relu'))
fmodel.add(kr.layers.Dense(32, activation='relu'))
fmodel.add(kr.layers.Dense(16, activation='relu'))
fmodel.add(kr.layers.Dense(1, activation='sigmoid'))
fmodel.summary()
fmodel.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
fmodel.fit(X,Y, epochs=1200, batch_size=128, verbose=1, shuffle=True, class_weight=None, sample_weight=None) # train
```
